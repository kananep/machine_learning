```python
import numpy as np
```

```python
high_var_array = np.array([1,100,200,300,4000,5000])
low_var_array = np.array([2,4,6,8,10])
```

```python
#variance
print(np.var(high_var_array))
np.var(low_var_array)
```

```python
#std
print(np.std(high_var_array))
np.std(low_var_array)
```

```python
#mean 
print(np.mean(high_var_array))
np.mean(low_var_array)
```

The standard deviation means that , with the high var array each number is 2072.711623024829 away from the mean (1600.1666666666667) and the with the low var arrar each number is 2.8284271247461903 away from the mean (6.0)

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.hist(high_var_array)
plt.show()
```

```python
heart = pd.read_csv('heart-disease.csv')
heart.head()
```

```python
heart.age.plot.hist()
```

```python
heart.plot.hist(figsize = (10, 30), subplots=True)
```

```python
heart.head()
```

```python
over_50 = heart[heart.age > 50]
over_50.head()
```

```python
over_50.plot(kind= 'scatter',
             x='age',
             y='chol',
             c = 'target')
```

```python
#OO method

fig , ax = plt.subplots(figsize=(10,6))
over_50.plot(kind= 'scatter',
             x='age',
             y='chol',
             c = 'target',
             ax =ax )

#ax.set_xlim([45 ,100])
```

```python
#OO method

fig , ax = plt.subplots(figsize=(10,6))

scatter = ax.scatter(x = over_50.age,
                     y= over_50.chol,
                     c = over_50.target,
                     )

ax.set(title='Heart and Cholesterol level',
       xlabel = 'Age',
       ylabel = 'Cholesterol')

ax.legend(*scatter.legend_elements(), title ='Target')
ax.axhline(over_50.chol.mean(),
           linestyle ='--')
```

```python
#subplot of chol, age , thalach

fig, (ax0,ax1) = plt.subplots(nrows =2 ,
                              ncols = 1,
                              figsize =(10 ,10))

scatter = ax0.scatter(x = over_50.age,
                     y= over_50.chol,
                     c = over_50.target,
                     )

ax0.set(title='Heart and Cholesterol level',
       xlabel = 'Age',
       ylabel = 'Thalach')
ax0.legend(*scatter.legend_elements(), title ='Target')
ax0.axhline(over_50.chol.mean(),
           linestyle ='--')

scatter = ax1.scatter(x = over_50.age,
                     y= over_50.thalach,
                     c = over_50.target,
                     )

ax1.set(title='Heart and Cholesterol level',
       xlabel = 'Age',
       ylabel = 'Thalach')
ax1.legend(*scatter.legend_elements(), title ='Target')
ax1.axhline(over_50.thalach.mean(),
           linestyle ='--')


fig.suptitle ('Heart Disease Analysis' , fontsize=16 , fontweight = 'bold')
```
